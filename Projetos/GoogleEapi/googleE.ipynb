```python
import googlemaps
import os
import dotenv
from pprint import pprint
import time
import pandas as pd
```

```python
%load_ext dotenv
```

```python
chave = os.getenv("API_CHAVE")
```

```python
map_client = googlemaps.Client(chave)

dir(map_client)
```

```python
nome_local = "Lojas Renner"
response = map_client.places(query = nome_local)
pprint(response)
```

```python
nome_sheet = 'Lista de Lojas | Stores List'
planilhaR = pd.read_excel("Planilhas e Fundamentos .xlsx", sheet_name = nome_sheet, skiprows = 5,
                         usecols = ['Business', 'Country', 'UF','State','City'])
planilhaR
```

```python
planilhaR = planilhaR.dropna()
planilhaR = planilhaR[(planilhaR['Business'] == "Renner") & (planilhaR['Country'] == "Brazil")]
planilhaR

```

```python
planilhaR['enderecos'] = (planilhaR[['City', 'State', 'UF' ,'Country']].apply(lambda row: '-'.join(row.values.astype(str)),axis=1))
enderecos_renner = list(planilhaR['enderecos'].unique())
enderecos_renner
```

```python
lista_coordenadas= []
for endereco in enderecos_renner: 
    response = map_client.geocode(endereco)

    latitude = response[0]['geometry']['location']['lat']
    longitude = response[0]['geometry']['location']['lng']

    coordenadas = (latitude, longitude)
    lista_coordenadas.append(coordenadas)
lista_coordenadas
```

```python
lista_dfs = []
for coordenadas in lista_coordenadas:
    print(coordenadas)
    try: 
        results = map_client.places_nearby(location= coordenadas, 
                                        keyword = "Lojas Renner", radius = 50000)
        while True:
            time.sleep(0.5)
            for store in results['results']:
                df = pd.DataFrame({'nome': [store['name']],
                                    'endereco': [store['vicinity']],
                                    'latitude': [store['geometry']['location']['lat']],
                                    'longitude': [store['geometry']['location']['lng']]})
                lista_dfs.append(df)
            if 'next_page_token' in results:
                next_page_token = results['next_page_token']
                results =map_client.places_nearby(location = coordenadas,
                                                keyword = "Lojas Renner", radius = 50000, page_token = next_page_token)
            else: 
                break
    except Exception as e:
        print(f"Erro ao obter dados: {e}")

lojas_renner = pd.concat(lista_dfs, ignore_index= True)
lojas_renner
```

```python
lojas_renner_filtrado = lojas_renner[(lojas_renner['nome'] == 'Renner') |
                                     (lojas_renner['nome'] == 'Lojas Renner') |
                                     (lojas_renner['nome'] == 'lojas Renner')
                                    ]
lojas_renner_filtrado = lojas_renner_filtrado.drop_duplicates(subset = ['endereco'])
```

```python
coordenadas_lojas = lojas_renner_filtrado[['latitude', 'longitude']].values
coordenadas_lojas
```

```python
import folium
from folium.plugins import HeatMap
```

```python
mapa = folium.Map(location= [-9.026078,-70.441312], zoom_start=6)
HeatMap(coordenadas_lojas, radius= 50).add_to(mapa)
mapa.save("heatmap_renner.html")
```
