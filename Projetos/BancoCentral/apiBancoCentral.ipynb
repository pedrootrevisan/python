```python
jump = 0
lista_dfs_moedas = []

while True:
    
    url_base = f'''https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=10000&$skip={jump}&$format=json'''

    response = requests.get(url_base)

    jump+=10000

    json_moedas = response.json()

    df_moedas = pd.DataFrame(json_moedas['value'])

    lista_dfs_moedas.append(df_moedas)

    if len(json_moedas['value']) <1:
        break


base_completa = pd.concat(lista_dfs_moedas)
base_completa
```
